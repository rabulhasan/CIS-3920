```r
# seq(0,1,length=10) makes a sequence of 10 numbers that are equally spaced between 0 and 1.
x=seq(-pi,pi,length=50)
x
```

```r
y=x
f=outer(x,y,function(x,y)cos(y)/(1+x^2))
#The contour() function produces a contour plot in order to represent three-dimensional data like a topographical map
contour(x,y,f)
contour(x,y,f,nlevels=45,add=T)
```

```r
fa=(f-t(f))/2
contour(x,y,fa,nlevels=15)
```

```r
image(x,y,fa)
```

The image() function works the same way as contour(), except that it produces a color-coded plot whose colors depend on the z value. This is known as a heatmap. 

```r
#persp() can be used to produce a three-dimensional plot
persp(x,y,fa)
```

```r
#The arguments theta and phi control the angles at which the plot is viewed
persp(x,y,fa,theta=30)
```
